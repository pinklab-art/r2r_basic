```python
import rclpy as rp
from turtlesim.msg import Pose
```

```python
rp.init()
```

```python
test_node = rp.create_node('sub_test')
```

```python
def callback(data):
    print("--->")
    print("/turtle1/pose : ", data)
    print("X : ", data.x)
    print("Y : ", data.y)
    print("Theta : ", data.theta)
```

```python
test_node.create_subscription(Pose, '/turtle1/pose', callback, 10)
```

```python
rp.spin_once(test_node)
```

```python
rp.spin(test_node)
```
